```python
import matplotlib.pyplot as plt                                                                     
import numpy as np
import math
from IPython.display import display, Markdown, clear_output, Math, Latex
import ipywidgets as widgets
from ipywidgets import Button, Layout
```

```python
def GetValidLS(j, parity, s1, p1, s2, p2):
    
    valid_l = []
    valid_s = []

    for S in range(abs(s1 - s2),(s1 + s2)+1):
        for L in range(abs(j - S),(j + S)+1):
            if(abs(p1*p2*pow(-1.,L) - parity) < 1.e-5):

                valid_l.append(L)
                valid_s.append(S)

    valid_ls = []
    valid_ls.append(valid_l)
    valid_ls.append(valid_s)
    
    return valid_ls;

def GetJPC(P_string):

    _pJ = int(P_string[:1])
    _pP = 0
    _pC = 0

    if("++" in P_string):
        _pP = 1
        _pC = 1
    
    if("+-" in P_string):
        _pP = 1
        _pC = -1

    if("-+" in P_string):
        _pP = -1
        _pC = 1

    if("--" in P_string):
        _pP = -1
        _pC = -1

    result = []
    result.append(_pJ)
    result.append(_pP)
    result.append(_pC)

    return result;


def JPCdecays(_motherJPC, _daughter1JPC, _daughter2JPC):

    _definedCParity = True
    
    bosons = False
    sameDaughters=False

    _allowedCParityLs = []
    _allowedCParitySs = []

    _forbiddenCParityLs = []
    _forbiddenCParitySs = []

    _undefinedCParityLs = []
    _undefinedCParitySs = []

    if(_daughter1JPC==_daughter2JPC):
        sameDaughters=True

    _motherJ = GetJPC(_motherJPC)[0]
    _motherP = GetJPC(_motherJPC)[1]
    _motherC = GetJPC(_motherJPC)[2]

    _daughter1J = GetJPC(_daughter1JPC)[0]
    _daughter1P = GetJPC(_daughter1JPC)[1]
    _daughter1C = GetJPC(_daughter1JPC)[2]
    
    _daughter2J = GetJPC(_daughter2JPC)[0]
    _daughter2P = GetJPC(_daughter2JPC)[1]
    _daughter2C = GetJPC(_daughter2JPC)[2]

    _allLSs=GetValidLS(_motherJ, _motherP, _daughter1J, _daughter1P, _daughter2J, _daughter2P)
    
    allowedC=True
    
    #if(sameDaughters==False):
    if (_motherC != (_daughter1C*_daughter2C)): 
        allowedC=False

    if(_motherC==0): 
        _definedCParity=False
   
    for i in range(0,len(_allLSs[0])):
        if(_definedCParity==True):
            if(sameDaughters==True):
                bosons = True
            if((2*_daughter1J)%2 !=0):
       
                bosons=False

                theL=_allLSs[0][i]
                theS=_allLSs[1][i]
                cparity=0

                if(bosons==True): 
                    cparity=pow(-1,theL)
                else: 
                    cparity=pow(-1,(theL+theS))

                if(_motherC != cparity ): allowedC=False

            if (allowedC==True): 
                _allowedCParityLs.append(_allLSs[0][i])
                _allowedCParitySs.append(_allLSs[1][i])

            else:
                _forbiddenCParityLs.append(_allLSs[0][i])
                _forbiddenCParitySs.append(_allLSs[1][i])
        else:
            _undefinedCParityLs.append(_allLSs[0][i])
            _undefinedCParitySs.append(_allLSs[1][i])

    if(len(_allowedCParityLs)>0):
        plt.scatter(_allowedCParityLs, _allowedCParitySs, color='crimson', s=16**2)
        xint = range(min(_allowedCParityLs), math.ceil(max(_allowedCParityLs))+1)
        yint = range(min(_allowedCParitySs), math.ceil(max(_allowedCParitySs))+1)
        plt.xticks(xint)
        plt.yticks(yint)
        plt.xlabel('L', fontsize=16)
        plt.ylabel('S', fontsize=16)
        plt.grid()
        
    if(len(_forbiddenCParityLs)>0):
        plt.scatter(_forbiddenCParityLs, _forbiddenCParitySs, color='dodgerblue', s=16**2)
        xint = range(min(_forbiddenCParityLs), math.ceil(max(_forbiddenCParityLs))+1)
        yint = range(min(_forbiddenCParitySs), math.ceil(max(_forbiddenCParitySs))+1)
        plt.xticks(xint)
        plt.yticks(yint)
        plt.xlabel('L', fontsize=16)
        plt.ylabel('S', fontsize=16)
        plt.grid()
    

    print("The following combinations are allowed for the decay:\n")

    if(_definedCParity==True):
        decaystring = '$\\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\;  \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\mathbf{' +str(_motherJPC)[:1]+ '^{' + str(_motherJPC)[1:-1] + str(_motherJPC)[2:] + '} \\rightarrow ' + str(_daughter1JPC)[:1] + '^{' + str(_daughter1JPC)[1:-1] + str(_daughter1JPC)[2:] + '} + ' + str(_daughter2JPC)[:1] + '^{' + str(_daughter2JPC)[1:-1] + str(_daughter2JPC)[2:] + '}}$'
        display(Latex(decaystring))
    else:
        decaystring = '$\\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\; \\mathbf{' +str(_motherJPC)[:1]+ '^{' + str(_motherJPC)[1:] + '} \\rightarrow ' + str(_daughter1JPC)[:1] + '^{' + str(_daughter1JPC)[1:] + '} + ' + str(_daughter2JPC)[:1] + '^{' + str(_daughter2JPC)[1:] + '}}$'
        display(Latex(decaystring))
    
    if(_definedCParity==True):
        if(len(_allowedCParityLs)+len(_forbiddenCParityLs)==0): 
            print("{: ^66s}".format('\33[1;41m'+"!!! NO VALID COMBINATION FOUND !!!\033[0m \n"))
        
        if(len(_allowedCParityLs)!=0 and allowedC==True):
            print("\n ________ Combinations with allowed C parity ________\n")
            for i in range(0,len(_allowedCParityLs)):
                print("{: ^66s}".format("\033[1;32;48m" + "L = " + str(_allowedCParityLs[i]) + ", S = " + str(_allowedCParitySs[i])+"\033[0m \n"))
                
        if(len(_forbiddenCParityLs)!=0 and allowedC==False):
            print("\n ________ Combinations with forbidden C parity ________\n")
            for i in range(0,len(_forbiddenCParityLs)):
                if(len(_forbiddenCParityLs)!=0):
                    print("{: ^66s}".format("\033[1;34;48m" +"L = " + str(_forbiddenCParityLs[i]) + ", S = " + str(_forbiddenCParitySs[i]) + "\033[0m \n"))

    else: 
        print("\n ________ C parity is not defined ________\n")
        print("\n ________ Allowed combinations ________\n")
        for i in range(0,len(_undefinedCParityLs)):
            if(len(_undefinedCParityLs)):
                print("{: ^66s}".format("\033[1;32;48m" + "L = " + str(_undefinedCParityLs[i]) + ", S = " + str(_undefinedCParitySs[i]) + "\033[0m \n"))
    
```

```python
l = Layout(flex='auto auto auto', height='30px', min_height='30px',  max_height='30px', width='150px', description_width= 'initial')
style = {'description_width': 'initial'}

mother = widgets.Textarea(
    value='0++',
    placeholder='JPC mother',
    description='$\mathrm{J^{PC}}$ mother',
    layout=l,
    style=style,
    disabled=False
)

daughter1 = widgets.Textarea(
    value='1--',
    placeholder='JPC daughter 1',
    description='$\mathrm{J^{PC}}$ daughter1',
    layout=l,
    style=style,
    disabled=False
)

daughter2 = widgets.Textarea(
    value='1--',
    placeholder='JPC daughter 2',
    description='$\mathrm{J^{PC}}$ daughter2',
    layout=l,
    style=style,
    disabled=False
)

def printer(mother, daughter1, daughter2): 
    print(JPCdecays(str(mother),str(daughter1),str(daughter2)))
```

### Specify the $\mathbf{J^{PC}}$ quantum numbers to be checked in the following cells for the decay:
### mother -> daughter1 + daughter2
### (in case no C parity is defined in the desired decay, just enter $\mathbf{J^{P}}$ like 1+ -> 2- + 1-)

```python
# hier nix ändern!
ui = widgets.HBox([mother, daughter1, daughter2])
out = widgets.interactive_output(printer,{'mother': mother, 'daughter1':daughter1, 'daughter2': daughter2})
display(ui, out)
plt.show()
# nur hier drunter in die Felder tippen!
```
